##cleanText

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 
# CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext

def cleanText(text, remove_spacing = False):
    if remove_spacing: return cleanhtml(text).replace('\n', ' ').replace('&nbsp;', '')
    else: return cleanhtml(text).replace('&nbsp;', '')
```

#Processing Class

```python
import re
import random
from tqdm.auto import tqdm
from multiprocessing import Pool
import multiprocessing

class prompt_class():
    def __init__(self, csv_file, output_dir = None):
        self.file = csv_file.split('/')[-1][:-4]

        #Set output directory
        if output_dir is None:
            self.output_dir = '/'.join(csv_file.split('/')[:-1])
        else:
            self.output_dir = output_dir

        self.value_map = {0:str(int(0)), 1:str(int(1)), 2:str(int(2)), 3:str(int(3)), 4:str(int(4))}

        self.df = self.process_problem_csv(csv_file)
        

    def process_problem_csv(self, csv_path):
        csv_df = pd.read_csv(csv_path)

        #Filter CSV
        # csv_df = csv_df[csv_df.problem_order==1]
        csv_df['correctness'] = csv_df['correctness'].map(lambda x: self.value_map[int(x*4)] if (x*4)%1 == 0 else np.nan)
        csv_df = csv_df.dropna(subset=['problem_body', 'answer_text', 'correctness', 'teacher_feedback'])

        #Format CSV
        csv_df['problem_body'] = csv_df['problem_body'].map(lambda x: cleanText(x, remove_spacing = True))
        csv_df['answer_text'] = csv_df['answer_text'].map(lambda x: cleanText(x, remove_spacing = True))
        csv_df['teacher_feedback'] = csv_df['teacher_feedback'].map(lambda x: cleanText(x, remove_spacing = True))

        return csv_df

    def construct_score_map(self, order):
        csv_df = self.df[self.df.problem_order == order]
        #Segregate CSV into map
        score_map = {}
        for score in csv_df['correctness'].unique():
            score_map[score] = csv_df[csv_df.correctness == score]

        return score_map

    def select_random_entry(self, df):
        length = df.shape[0]
        num = int(length * random.random())
        return df.iloc[num]

    def get_prompt(self, csv_df_iloc, task = '', include_question = True, include_answer=True, include_feedback = False, include_score = True):
        prompt_text = task if len(task)>0 else ''
        
        if include_question: 
            QUESTION = "QUESTION: " + csv_df_iloc['problem_body'] + "\n"
            prompt_text += QUESTION

        if include_answer:
            ANSWER = "ANSWER: " + csv_df_iloc['answer_text'] + "\n"
            prompt_text += ANSWER

        if include_score:
            SCORE = "SCORE: " + str(csv_df_iloc['correctness']) + "\n"  
        # else: SCORE = "What score should this answer have?  "
        else: SCORE = "SCORE: "
        prompt_text += SCORE

        if include_feedback: 
            FEEDBACK = "FEEDBACK: " + csv_df_iloc['teacher_feedback'] + "\n\n"
            prompt_text += FEEDBACK

        return prompt_text

    def get_question(self, part):
        df = self.df

        problem_body = ''
        for i in range(part):
            p = i+1
            problem_body += df[df.problem_order==p]['problem_body'].unique()[0] + '\n'
            print(problem_body)

        return problem_body


    def construct_random_prompt(self, score_map, selected_problem, num_groups=1):
    
        keys = list(score_map.keys())
        task = 'TASK: Classify the Score of the following question as 0, 1, 2, 3, or 4.\n'

        # #Get Problem Body
        # problem_body = 

        task += "QUESTION: " + selected_problem['problem_body'] + "\n\n"

        for i in range(num_groups):
            random.shuffle(keys)
            for i in keys:
                sample_not_found = True
                sample_attempt_count = 0
                while sample_not_found:
                    sample_attempt_count += 1
                    sample = self.select_random_entry(score_map[i])
                    if (sample.problem_log_id != selected_problem.problem_log_id): 
                        sample_not_found = False 
                    if sample_attempt_count == 5:
                        return
                    

                task = self.get_prompt(sample, task, include_question = False, include_feedback = True)

        final_prompt = self.get_prompt(selected_problem, task, include_question = False, include_score=False)
        # print(final_prompt)
        return final_prompt


    def make_dataset(self, num_groups = 1, num_prompts = 1, multi_processing = False):
        df = self.df
        value_map = self.value_map
        output_dir = self.output_dir
        output = {
            'problem_log_id': [],
            'student_user_id': [],
            'teacher_id':[],
            'problem_order': [],
            'prompt': [],
            'correctness': [],
            'teacher_feedback': []
        }

        if multi_processing == False:
            for index in tqdm(range(df.shape[0])):
                # entry = entry[1]
                entry = df.iloc[index]
                order = entry['problem_order']
                ScoreMap = self.construct_score_map(order)
                for np in range(num_prompts):
                    prompt = self.construct_random_prompt(ScoreMap, entry, num_groups)

                    output['problem_log_id'] += [entry['problem_log_id']]
                    output['student_user_id'] += [entry['student_user_id']]
                    output['teacher_id'] += [entry['teacher_id']]
                    output['problem_order'] += [order]
                    output['prompt'] += [prompt]
                    output['correctness'] += [entry['correctness']]
                    output['teacher_feedback'] += [entry['teacher_feedback']]
                
                csv_name = f"{self.file}_{num_groups}_{num_prompts}.csv"
                file = os.path.join(self.output_dir, csv_name)
                output_csv = pd.DataFrame(output).to_csv(file)
        
        else:
            num_processes = multiprocessing.cpu_count()-1
            p = Pool(processes=num_processes)

            chunk_size = int(df.shape[0]/num_processes)
            chunks = [df.iloc[i:i + chunk_size] for i in range(0, df.shape[0], chunk_size)]

            def process_problem(entry):
                entry = entry[1]
                order = entry['problem_order']
                ScoreMap = self.construct_score_map(order)

                output = {
                            'problem_log_id': [],
                            'student_user_id': [],
                            'teacher_id':[],
                            'problem_order': [],
                            'prompt': [],
                            'correctness': [],
                            'teacher_feedback': []
                        }

                for np in range(num_prompts):
                    prompt = self.construct_random_prompt(ScoreMap, entry, num_groups)

                    output['problem_log_id'] += [entry['problem_log_id']]
                    output['student_user_id'] += [entry['student_user_id']]
                    output['teacher_id'] += [entry['teacher_id']]
                    output['problem_order'] += [order]
                    output['prompt'] += [prompt]
                    output['correctness'] += [entry['correctness']]
                    output['teacher_feedback'] += [entry['teacher_feedback']]

                pbar.update(1)
                return output

            output_csv = {}
            results = p.map(self.process_problem, chunks)
            for r in results:
                output_csv.update(r)

            csv_name = f"{self.file}_{num_groups}_{num_prompts}.csv"
            file = os.path.join(self.output_dir, csv_name)
            output_csv = pd.DataFrame(output).to_csv(file)

            p.close()
            p.join()
            pbar.close()
        
            
        return(output_csv)
    
```

##set up model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

device = 'cpu'

tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-560m")

model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-560m")

input_sentence = "The SQL command to extract all the users whose name starts with A is: "
```

##run_model

```python
def run_model(input, model, tokenizer, max_tokens = 1, verbose = False):
    input_ids = tokenizer.encode(input, return_tensors='pt').to(device)
    model.to(device)

    for i in range(max_tokens): 
        translation = model(input_ids)
        mask = torch.zeros_like(translation.logits[:, -1, :])
        mask.T[19:23] = 1
        # output = translation.logits[:, -1, :] * mask
        output = translation.logits[:, -1, :]
        translated = torch.argmax(output, dim=-1)

        input_ids = torch.cat([input_ids, translated.unsqueeze(-1)], dim=-1)

    if verbose: print(tokenizer.decode(input_ids.squeeze(), skip_special_tokens=True))

    translation = model(input_ids)
    
    
    output_text = tokenizer.batch_decode(translated, skip_special_tokens=True)[0]
    # print(translated_text)
    return output_text

run_model(input_sentence, model, tokenizer)
```
